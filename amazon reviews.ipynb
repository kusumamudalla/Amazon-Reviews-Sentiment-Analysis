7.Encoding Categorical Values

```python

from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()


df['sentiments']= label_encoder.fit_transform(df['sentiments'])

df['sentiments'].unique()
```

```python
le_name_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print(le_name_mapping)
```

```python
df.head()
```

```python
X = df['cleaned_review'].values
y = df['sentiments'].values
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
nltk.download("stopwords")
```

```python
print(stopwords.words('english'))
```

```python
ps = PorterStemmer()

corpus = []

for i in range(len(X)):

    review = re.sub('[^a-zA-Z]', ' ', X[i])
    review = review.lower()
    words = review.split()

    filtered_words=[]
    for word in words:
        if word not in stopwords.words("english"):
            word=ps.stem(word)
            filtered_words.append(word)

    tweet = ' '.join(filtered_words)

    corpus.append(review)
```

```python
X[0], corpus[0]
```

9.Splitting The Dataset Into Training And Test Set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(corpus,y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.feature_extraction.text import  TfidfVectorizer
# from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from imblearn.over_sampling import RandomOverSampler
from imblearn.pipeline import Pipeline
```

```python
clf = SVC(random_state=42)
sampler = RandomOverSampler(random_state=0)
tfidf = TfidfVectorizer()
```

```python
pipe_line = Pipeline( [ ('tfidf', tfidf), ('RandomOverSampling',sampler), ('model', clf) ] )
```

```python
pipe_line.fit(X_train, y_train)
```

```python
y_predit=pipe_line.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_predit)) 
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,y_predit)
print(accuracy)
```

Save Model Using Pickel

```python
import pickle
pickle.dump(pipe_line,open("pipeline.pickle","wb"))
```
